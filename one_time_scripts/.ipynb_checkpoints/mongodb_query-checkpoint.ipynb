```python
import pymongo
import json
from bson import BSON
import pandas as pd
import os
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
db_client = pymongo.MongoClient("mongodb://10.100.74.70:27017/")
```

```python
db_client.list_database_names()
```

```python
articles_db = db_client.Articles
```

```python
articles_db
```

```python
articles_db.list_collection_names()
```

```python
lemonde_col = articles_db['lemonde']
```

```python
lemonde_col.count_documents({})
```

```python
lemonde_col.count_documents(filter={"article_content.author_name":  "Le Monde"})
```

```python
lemonde_col.count_documents(filter={"article_content.title":  {
    '$regex': 'Macron'
} })
```

```python
results = lemonde_col.find({"article_content.title":  {
                     '$regex': 'Macron'
                     } 
                 },
                 {"publication_date": 1, 'article_content.title': 1, 'topic_category': 1}
                ).sort("publication_date", 1)
```

```python
for item in results:
    print(item)
```

```python
results_df = pd.DataFrame.from_records(results)
```

```python
results_df
```

```python
results_df.dtypes
```

```python
# plt.figure(figsize=(30, 10))
results_df[results_df['publication_date'] < datetime(year=2016, month=1, day=1)]['topic_category'].value_counts()
```

```python
results_df[results_df['publication_date'] < datetime(year=2016, month=1, day=1)].shape
```
